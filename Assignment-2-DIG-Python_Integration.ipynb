```python
import mysql.connector
from mysql.connector import Error
import pandas as pd
```

```python
global connection,cur
```

```python
# create connection
def create_server_connection(host_name,user_name,password):

    
    connection = None

    try:
        connection = mysql.connector.connect(
            host = host_name,
            user = user_name,
            password = password)
        print("mysql connection successfull")

    except Error as er:
        print(f"Error: {er}")

    return connection

create_server_connection("localhost","root","1234")
```

```python
# create database
def create_database(connection,query):
    cur = connection.cursor()

    try:
        cur.execute(query)
        print("Database created")

    except Error as er:
        print(f"Error : {er}")

create_db_query = "create database Internship"

create_database(connection,create_db_query)
```

```python
# connecting to database

def create_db_connection(host_name,user_name,pswd,db):
    connection = None
    try:
        connection = mysql.connector.connect(host = host_name, user = user_name, password = pswd, database = db)
        print("Database connection successfully")

    except Error as er:
        print(f"Error: {er}")
    
    return connection

create_db_connection("localhost","root","1234","Internship")
```

```python
def execution(connection,query):
    cur = connection.cursor()
    try:
        cur.execute(query)
        connection.commit()
        print("Query executed successfully")
    except Error as er:
        print(f"Error: {er}")
```

```python
create_orders_table = """
Create table orders(
order_id int primary key,
customer_name varchar(30) not null,
product_name varchar(30) not null,
date_ordered date,
quantity int,
unit_price float,
phone_number varchar(30));
"""
create_db_connection("localhost","root","1234","Internship")
# execution(connection,"use internship")
execution(connection,create_orders_table)
```

```python
def show_query(connection,query):
    cur = connection.cursor()
    output = None
    try:
        cur.execute(query)
        output = cur.fetchall()
        return output
    except Error as er:
        print(f"Error: {er}")

select = "select * from orders"
output = show_query(connection,select)
```

```python

for i in output:
    print(i)
```

```python
new_df = []

for i in output:
    new_df.append(list(i))

columns = ["order_id", "customer_name", "product_name", "date_ordered", "quantity",
"unit_price", "phone_number"]

df = pd.DataFrame(new_df,columns= columns)
df
```
